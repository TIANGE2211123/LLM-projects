```python
import numpy as np
import matplotlib.pyplot as plt

# 生成 e 的取值范围 (0 到 1)
e_values = np.linspace(0, 1, 500)

# 设置不同的 d 值
d_values = [0, 0.2, 0.5, 0.8]

# 创建图表
plt.figure(figsize=(10, 6))

# 为每个 d 值绘制一条曲线
for d in d_values:
    speedup_values = 1 / (e_values / 20 + d / 10 + (1 - e_values - d))
    
    # 检查并标记无穷大的情况
    infinite_points = np.isinf(speedup_values)
    if any(infinite_points):
        print(f"Infinite speedup found for d = {d} at e = {e_values[infinite_points]}")
    
    plt.plot(e_values, speedup_values, label=f'd = {d}')

# 添加横坐标焦点
focal_points = [0, 0.5, 1]  # 可以修改为其他感兴趣的值
for point in focal_points:
    plt.axvline(x=point, color='gray', linestyle='--', linewidth=0.7)
    plt.text(point, plt.ylim()[1]*0.95, f'e={point}', color='black', fontsize=10, ha='center')

# 设置图表属性
plt.xlabel('e')
plt.ylabel('y')
plt.title('y - e')
plt.grid(True)
plt.legend()
plt.show()
```
